```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
m = 0.43 # [kg]
x0 = 0  #[m]
y0 = 0  #[m]
vx0 = 15*np.sqrt(3) #[m/s]
vy0 = 15 # [m/s]
g = 9.81 # [m/s^2]

```

```python
dt = 0.001 #[s]
%matplotlib notebook
t = np.arange(0, 3.05, dt)
```

```python
b = 0.006 # [Ns/m]
t = np.arange(0, 3.05, 0.01)
```

```python
x = x0
y = y0
vx = vx0
vy = vy0
r = np.array([x,y])
while y >= 0:
    dxdt = vx
    x = x + dt * dxdt
    dydt = vy
    y = y + dt * dydt
    dvxdt = -b/m*np.sqrt(vx**2+vy**2)*vx
    vx = vx + dt * dvxdt
    dvydt = -b/m*np.sqrt(vx**2+vy**2)*vy - g
    vy = vy + dt * dvydt
    r = np.vstack((r, np.array([x,y])))

plt.figure()
plt.plot(r[:,0],r[:,1])
plt.xlim(0,80)
plt.ylim(0,12)
plt.title(' Exercício Modelo')
plt.xlabel('x (m)', size = 20)
plt.ylabel('y (m)', size = 20)
plt.grid()
plt.show()
```
